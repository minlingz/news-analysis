```python
import json
import pandas as pd


def extract_summary(file, n=10):
    with open(file) as f:
        data = json.load(f)
    # Initialize a list to store the extracted values
    summary = []
    for item in data:
        s = {}
        if "summary" in item:
            s["summary"] = item["summary"]
            summary.append(s)
    return summary[:n]


# read the summary from json file
summary = []
file_list = pd.read_csv(
    "/Users/minlingzhou/Library/CloudStorage/OneDrive-DukeUniversity/vc/Summer/NarrativeNexus/data/meta_data/QnA_file_list.csv"
)
for file in file_list["file_name"]:
    s = extract_summary(
        "/Users/minlingzhou/Library/CloudStorage/OneDrive-DukeUniversity/vc/Summer/NarrativeNexus/data/QnA_data/"
        + file
    )
    summary = summary + s
```

```python
from dotenv import load_dotenv
from openai import OpenAI
import os

load_dotenv(override=True)  # This loads the environment variables from .env file
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
def QnA(
    summary,
    client,
    model,
    QnA_prompt,
):

    prompt_complete = (
        QnA_prompt + "\n" + "Here are the news articles summaries:" + "\n" + summary
    )
    model = model
    messages = [{"role": "user", "content": prompt_complete}]
    try:
        response = client.chat.completions.create(
            model=model, messages=messages, temperature=0
        )

        content = response.choices[0].message.content
        return content
    except Exception as e:  # if the model fails to return a response
        print(f"Error: {e}")
        return "Sorry, error from GPT."
```

```python
prompt = "I have a collection of 60 news articles covering a wide range of topics. I need you to propose 8 multiple-choice questions that can help in understanding the similarities and differences among these topics. Each question should focus on specific aspects of the articles such as the type of sources used, main topics, geographical focus, key issues, tone, intended audience, format, key figures, primary objective, and supporting evidence. \n \
    Here are the guidelines for creating the questions: \n \
       1.	Identify Key Topics: Begin by identifying the primary topics and subtopics covered in the articles. Group articles with similar themes together. \n \
       2.	Determine Commonalities and Differences: Look for common elements and significant differences within and across the topic groups. Focus on aspects such as key events, figures, locations, and underlying issues. \n \
       3.	Formulate Questions: Create questions that prompt readers to identify these commonalities and differences. Ensure each question has one correct answer and multiple plausible distractors. \n \
    Example Question Format: \n \
    •   Question: Clearly state the question, focusing on a specific aspect of the articles. \n \
	•	Choices: Provide four or five answer choices (A, B, C, D, E), with one correct answer and several plausible distractors. \n \
	•	Explanation (optional): Offer a brief explanation of the correct answer, highlighting the relevant similarities or differences. \n \
    Example Questions: \n \
    Question: What is the main topic of the article? \n \
    •	A) Politics and Governance \n \
	•	B) Environment and Climate Change \n \
	•	C) Health and Medicine \n \
	•	D) Technology and Innovation \n \
	•	E) None of the above"
```

```python
response = QnA(str(summary), client, "gpt-4o", QnA_prompt=prompt)
```
